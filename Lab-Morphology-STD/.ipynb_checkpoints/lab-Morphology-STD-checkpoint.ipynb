```python
from commonfunctions import *
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import skimage
from skimage.color import rgb2gray
from skimage.morphology import binary_erosion, binary_dilation, binary_closing,skeletonize, thin
from skimage.measure import find_contours

from skimage.draw import rectangle
from skimage.draw import rectangle_perimeter
from skimage.filters import threshold_otsu
# Show the figures / plots inside the notebook
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python


def binarize(img,thresh=None):
    img_cp = img.copy()
    threshold = threshold_otsu(img_cp) if thresh is None else thresh
    img_cp[img_cp < threshold] = 0
    img_cp[img_cp >= threshold] = 255
    return img_cp

def erode_image(img, kernel=np.ones((2,2))):
    kernel = kernel*255
    img_cp = img.copy()
    h = len(img)
    w = len(img[0])
    k_sz = len(kernel)

    for i in range(k_sz, h - k_sz // 2):
        for j in range(k_sz, w - k_sz //2):
            identical = True
            for k in range(k_sz):
                for m in range(k_sz):
                    if(kernel[k][m] != img[i-k][j-m]):
                        identical = False
                        break
                    if not identical:
                        break
            if identical:
                img_cp[i][j] = 255
            else:
                img_cp[i][j] = 0
                
    return img_cp

def dilate_image(img, kernel=np.ones((2,2))):
    kernel = kernel*255
    img_cp = img.copy()
    h = len(img)
    w = len(img[0])
    k_sz = len(kernel)

    for i in range(k_sz, h - k_sz // 2):
        for j in range(k_sz, w - k_sz //2):
            intersection = False
            for k in range(k_sz):
                for m in range(k_sz):
                    if(kernel[k][m] == img[i-k][j-m]):
                        intersection = True
                if intersection:
                    break
            if intersection:
                img_cp[i][j] = 255
            else:
                img_cp[i][j] = 0
                
    return img_cp


```

```python
'''
3) Skeletonization:
    - Load "horse.jpg" from the img folder 
    - Use a binary threshold of 0.9 to remove invisible noise in the background
    - Use Skimage's "skeletonize(image)" to acquire the image's skeleton & show it
    - Use Skimage's "thin(image, max_iter)" with values of max_iter 5/10/15/20 & show it
'''
horse = io.imread('img/horse.jpg')
horse = np.round(rgb2gray(horse)*255)
horse = horse.astype(np.uint8)
# horse_bin = binarize(horse)
horse_bin = binarize(horse,.9*255)
horse_skeleton = skeletonize(horse_bin/255)
show_images([horse_bin, horse_skeleton], ['binary image','horse skeleton'])
iterations = [5,10,15,20]
thinned = []
labels = []
non_closed_horse = horse
print("thinnnig before closing")
for it in iterations:
    img = thin(non_closed_horse, it)
    thinned.append(img)
    labels.append(f"thinned with max_it = {it}")
show_images(thinned, labels)

closed_horse = binary_closing(horse)
thinned = []
labels = []
print("thinnnig after closing")
for it in iterations:
    img = thin(closed_horse, it)
    thinned.append(img)
    labels.append(f"thinned with max_it = {it}")
show_images(thinned, labels)
```
