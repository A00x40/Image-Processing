```python
'''
(2) Median Filter
'''
def median_filter( data , window ):
    rows = len(data)
    columns = len(data[0]) 
    
    index = window // 2
    data_filtered = np.zeros( (rows,columns) )
    
    for x in range(rows) :
        
        for y in range(columns) :
            
            medianArray = np.zeros( (window,window) )
            
            for fx in range( window ) :
                for fy in range( window ) :
                    if x + fx - index >= 0 and x + fx - index <= rows-1 and y + fy - index >= 0 and y + fy - index <= columns-1 :
                        medianArray[fx][fy] = data[x + fx - index][y + fy - index]
                        
            medianArray.sort()
            data_filtered[x][y] = medianArray[window//2][window//2]
      
    return data_filtered
```

```python
pout = io.imread("pout.tif")
grayPout = rgb2gray(pout)
noisePout = random_noise(grayPout ,mode="s&p",amount = 0.05)
show_images([pout , grayPout  , noisePout] , ['Original','GrayScale','Noise'])

my_filtered = median_filter( noisePout , 3 )

skimed = median( noisePout )
show_images([my_filtered , skimed] , ['My Filtered','Skimage Filtered'])
```
