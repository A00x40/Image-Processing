```python
from commonfunctions import *
%matplotlib inline
%load_ext autoreload
%autoreload 2
from skimage.filters import roberts, sobel,sobel_h,sobel_v, prewitt,gaussian
import skimage.io as io
from scipy import ndimage
```

```python
'''
1. Built-in Sobel, Prewill, Roberts
2. Custom Sobel (V-H-Both)
3. Custom LoG
'''
```

```python
# 1. Built-in Sobel, Prewill, Roberts,canny

img = rgb2gray(io.imread('circuit.tif'))
edge_rob = roberts(img)
edge_pre = prewitt(img)
edge_sobelh = sobel_h(img)
edge_sobelv = sobel_v(img)
edge_sobel = sobel(img)
fig, ax = plt.subplots(ncols=3)
ax.ravel()
ax[0].imshow(img,cmap='gray')
ax[0].set_title("original")
ax[1].imshow(edge_rob,cmap='gray')
ax[1].set_title("robert")
ax[2].imshow(edge_pre,cmap='gray')
ax[2].set_title("prewitt")
fig, ax = plt.subplots(ncols=3)
ax.ravel()
ax[0].imshow(edge_sobelh,cmap='gray')
ax[0].set_title("sobel_h")
ax[1].imshow(edge_sobelv,cmap='gray')
ax[1].set_title("sobel_v")
ax[2].imshow(edge_sobel,cmap='gray')
ax[2].set_title("sobel")
fig, ax = plt.subplots(ncols=2)
edge_canny = canny(img,sigma=2)
edge_log=ndimage.gaussian_laplace(img, sigma=2.5)
ax[0].imshow(edge_canny,cmap='gray')
ax[0].set_title("canny")
ax[1].imshow(edge_log,cmap='gray')
ax[1].set_title("LOG")
```

```python
# 2. Custom Sobel (V-H-Both)

hx=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
hy=hx.T

c,r=img.shape

#sobel_v
sobelx=[np.sum(hx*img[i-1:i+2,j-1:j+2]) for i in range(c-1) for j in range(r-1) if i>0 and j>0]
sobel_h
sobely=[np.sum(hy*img[i-1:i+2,j-1:j+2]) for i in range(c-1) for j in range(r-1) if i>0 and j>0]

sobelx=np.reshape(sobelx,(c-2,r-2))
sobely=np.reshape(sobely,(c-2,r-2))

#sobel
sobel=np.sqrt(sobelx*sobelx+sobely*sobely)

#thresholding
sobel[sobel<110]=0
sobel[sobel>=110]=255

sobelx[sobelx<110]=0
sobelx[sobelx>=110]=255

sobely[sobely<110]=0
sobely[sobely>=110]=255

#show
fig, ax = plt.subplots(ncols=4)
ax.ravel()
ax[0].imshow(img,cmap='gray')
ax[1].imshow(sobelx,cmap='gray')
ax[2].imshow(sobely,cmap='gray')
ax[3].imshow(sobel,cmap='gray')
```

```python
# 3. Custom LoG
thres1= 5 #thres f1
thres2=2  #thres f2
#gaussian filter
img_g= gaussian(img, sigma=2,preserve_range=True)

f1=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
f2=np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
c,r=img_g.shape

#LOG with f1
log1=[np.sum(f1*img_g[i-1:i+2,j-1:j+2]) for i in range(c-1) for j in range(r-1) if i>0 and j>0]
b1=[i for i in range(len(log1)) if log1[i]<=thres1]
w1=[i for i in range(len(log1)) if log1[i]>thres1]
log1=np.array(log1)
log1[b1]=0
log1[w1]=255

#LOG with f2
log2=[np.sum(f2*img_g[i-1:i+2,j-1:j+2]) for i in range(c-1) for j in range(r-1) if i>0 and j>0]
b2=[i for i in range(len(log2)) if log2[i]<=thres2]
w2=[i for i in range(len(log2)) if log2[i]>thres2]
log2=np.array(log2)
log2[b2]=0
log2[w2]=255

log1=np.reshape(log1,(c-2,r-2))
log2=np.reshape(log2,(c-2,r-2))

#show
fig, ax = plt.subplots(ncols=3)
ax.ravel()
ax[0].imshow(img_g,cmap='gray')
ax[1].imshow(log1,cmap='gray')
ax[2].imshow(log2,cmap='gray')
```
